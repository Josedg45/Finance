```python
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
```

```python
start = dt.datetime(2023,1,1)#year,month,day
end = dt.datetime(2024,1,1) #dt.datetime.now()
x = '^GSPC'
```

```python
data = yf.download(x, start=start, end=end)
```

```python
data = data[['Open','High','Low','Close']]
```

```python
data = data.droplevel(1, axis=1)
```

```python
mpf.plot(data, type='candle', style='yahoo', title=f'Stock Price of {x}', ylabel='Price ($)')
```
